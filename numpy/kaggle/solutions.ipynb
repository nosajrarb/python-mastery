```python
# 13. Create a 10x10 array with random values and find the minimum and maximum values
rng = np.random.default_rng()
random_10x10 = rng.random(size = (10,10)) *100
print('array: ', random_10x10)
min_val = np.amin(random_10x10)
max_val = np.amax(random_10x10)
print('min value: ',min_val)
print('max value: ', max_val)
```

```python
# 14. Create a random vector of size 30 and find the mean value
random_number_generator = np.random.default_rng()
randomx30 = random_number_generator.random(size=(30)) * 10
print('random size 30 vector: ', randomx30)
mean = np.mean(randomx30)
print('mean: ' , mean)
```
